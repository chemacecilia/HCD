1.     Mostrar la diferencia entre diferentes datums *horizontales* (como WGS84 y NAD83), al transformar un punto geográfico entre estos datums.
2.   Visualizar el desplazamiento en las coordenadas que resulta del cambio de datum, lo cual es fundamental para entender cómo la selección de un datum puede afectar la precisión y la ubicación en los mapas.


---
**Datum Horizontal:** Un sistema de referencia que define la posición de puntos en la Tierra. Se basa en una representación matemática de la forma de la Tierra, que puede variar según el datum.

**Ejemplos Comunes:** WGS84 (utilizado por GPS), NAD83 (usado en Norteamérica), ED50 (Datum Europeo).

**Ejemplo:** Transformación de Coordenadas entre Datums






1. **Instalación de Librerías**: Usaremos la librería pyproj para realizar las transformaciones de datum.

```python
# !pip install pyproj
```

```python
from pyproj import Transformer
import matplotlib.pyplot as plt
```

2. **Definir Puntos y Datums**
	* Definir un punto en coordenadas geográficas (latitud y longitud).

```python
# Definir el punto en WGS84 (por ejemplo, una ubicación en Valencia)
latitude = 39.47391
longitude = -0.37966
```

3. **Transformar este punto de WGS84 a NAD83 y mostrar la diferencia.**


```python
# Crear un transformador de WGS84 a NAD83
transformer = Transformer.from_crs("EPSG:4326", "EPSG:4230", always_xy=True)  # EPSG:4326 = WGS84, EPSG:4269 = NAD83

# Transformar el punto
longitude_nad83, latitude_nad83 = transformer.transform(longitude, latitude)

# Mostrar resultados
print(f"Coordenadas en WGS84: Latitud = {latitude}, Longitud = {longitude}")
print(f"Coordenadas en NAD83: Latitud = {latitude_nad83}, Longitud = {longitude_nad83}")
```

4. **Visualización del desplazamiento**


```python
plt.figure(figsize=(8, 6))
plt.plot(longitude, latitude, 'bo', label="WGS84 (Original)")
plt.plot(longitude_nad83, latitude_nad83, 'ro', label="NAD83 (Transformado)")
plt.legend()
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.title("Desplazamiento entre WGS84 y NAD83")
plt.grid(True)
plt.show()
```
